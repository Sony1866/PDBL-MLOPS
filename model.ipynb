```python
import pandas as pd
from pycaret.regression import *

# 1. Load data
print("Loading data...")
df = pd.read_csv("D:\\Koding Liburan\\Full Dataset\\cleaning_prosperloadata.csv")

# (Opsional) Drop kolom yang bisa bikin bias atau ga relevan buat diprediksi
# Misalnya kolom ID, atau status yang baru ketahuan setelah pinjaman jalan
cols_to_ignore = ['LoanStatus', 'ListingCategory (numeric)'] 

# 2. Setup PyCaret (Proses split data 80/20 & Preprocessing)
print("Setting up PyCaret...")
reg_setup = setup(data=df, 
                  target='BorrowerAPR', # Target prediksi kita
                  train_size=0.8, 
                  session_id=123, 
                  normalize=True, 
                  ignore_features=cols_to_ignore)

# 3. Model Selection (Sesuai request: XGBoost, LightGBM, Random Forest)
print("Training 3 Models...")
xgb_model = create_model('xgboost')
lgbm_model = create_model('lightgbm')
rf_model = create_model('rf')

# Bandingkan performa ketiga model secara otomatis
print("Comparing Models...")
best_model = compare_models(include=['xgboost', 'lightgbm', 'rf'])

# 4. Hyperparameter Tuning untuk XGBoost
print("Tuning XGBoost...")
tuned_xgb = tune_model(xgb_model)

# 5. Visualisasi Evaluasi Model (buat di-screenshot ke presentasi)
# Ini bakal ngeluarin grafik-grafik yang diminta di poin "Model Evaluation"
evaluate_model(tuned_xgb) # UI Interaktif di Jupyter
# Atau panggil spesifik:
# plot_model(tuned_xgb, plot='learning') # Learning curve
# plot_model(tuned_xgb, plot='error')    # Prediksi vs Aktual

# 6. Finalisasi & Save Model untuk Deployment
final_xgb = finalize_model(tuned_xgb)
save_model(final_xgb, 'model_xgboost_prosper')
print("Model berhasil disimpan untuk dideploy!")
```
